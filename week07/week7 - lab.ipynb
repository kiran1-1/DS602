```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/ds602/log_reg/employee-turnover-balanced.csv')
y = df['left_company']
X = df.iloc[:, 1:]
```

```python
X.shape
```

```python
y
```

```python
numerical_columns  = [column_name for column_name in X.columns if X[column_name].dtype!=object]
numerical_columns
```

```python
Categorical_columns  = [column_name for column_name in X.columns if X[column_name].dtype==object]
Categorical_columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipelinesd = Pipeline([#simple imputer helps in filling missing values with median 
    ('scaler', StandardScaler())# this line helps in standardizing values using a standard scalar
])
X_train_PL = pd.DataFrame(pipelinesd.fit_transform(X[numerical_columns]))
X_train_PL 
```

```python
from sklearn.preprocessing import OneHotEncoder
pipeline_Categorical_data = Pipeline([# using simple imputer for filling missing values with most_frequent
    ('encoder', OneHotEncoder(handle_unknown='ignore'))#using one_hot_encoder for converting categorical data into numeric data.
])
X_train_Catgorical_data = pipeline_Categorical_data.fit_transform(X[Categorical_columns])
data_ = pd.DataFrame(X_train_Catgorical_data.toarray())
data_
```

```python
combined_data = pd.concat([X_train_PL, data_], axis=1)# combining both numric and categorical data 
combined_data
```

```python
y.shape
```

Train and Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(combined_data, y, test_size=0.25, random_state=42)
```

Train a Decission tree model 

```python
from sklearn.tree import DecisionTreeClassifier

Decision_tree = DecisionTreeClassifier()
Decision_tree.fit(X_train, y_train)
```

Train a Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
```

Evaluate the models

```python
from sklearn.metrics import accuracy_score

decision_accuracy = accuracy_score(y_test, Decision_tree.predict(X_test))
random_forest_accuracy = accuracy_score(y_test, random_forest.predict(X_test))

print('Decision tree accuracy:', decision_accuracy)
print('Random forest accuracy:', random_forest_accuracy)
```

From above results we can say that random forest are better at predicting turnovers than decission tree why because :
    
Decision Trees can be prone to overfitting, especially when they are deep. Random Forests, by aggregating predictions from multiple trees, reduce overfitting and provide more generalizable models.

Due to the averaging effect of multiple trees, Random Forests  provided as with more accurate predictions, especially when the individual trees in the ensemble are diverse and not highly correlated.

Random Forests can provide a feature importance score, indicating the contribution of each feature in the prediction. This can be valuable for feature selection and understanding which variables are significant in making predictions.

Decission tree cannnot learn the features  whereas random forest can before fiting the model.

However, it's essential to note that there are cases where a single Decision Tree might be preferred, especially when interpretability is crucial, and you need a simple model that can be easily visualized and explained. The choice between Random Forests and Decision Trees depends on the specific problem, the complexity of the data, and the trade-off between accuracy and interpretability.
