## Question 4. Analyze the relationship between the wholeprice `mmr` and the `salesprice`.
- What is the overall correlation?  
- What happens to the correlation as you increase the maximum selling price window (e.g., 0-10000, 0-20000,...)?
- Create a scatter plot with the relationship.  
- Create a graph that shows how the correlation changes.

````python
#previosly while converting datatype i have assigned non-integer values to 0 so we remove the rows where mmr is equal to zero
dataframe_without_zeros = df2[df2['mmr'] != 0]
````

````python
#finding correlation using corr
print(dataframe_without_zeros['mmr'].corr(dataframe_without_zeros['sellingprice']))
````

````python
#creating a copy of a dataframe 
dataframe_for_Correlation = dataframe_without_zeros.copy()
#creating a selling price range window  for finding different correlation
Selling_price_window = [10000,20000,30000,40000]
#creating a empty list for storing correlation value
correlation_list = []
#looping through the selling prices and finding correlation for every range value
for val in Selling_price_window:
    dataframe_for_Correlation = dataframe_without_zeros[dataframe_without_zeros["sellingprice"] <= val]
    correlation_val = dataframe_for_Correlation['mmr'].corr(dataframe_for_Correlation['sellingprice'])
    correlation_list.append(correlation_val)
print(correlation_list)
````

````python
# scatter plot for mmr column and sellingprice column
plt.scatter(dataframe_without_zeros['mmr'], dataframe_without_zeros['sellingprice'], c='blue', marker='o', label='Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot for Two Columns')
plt.legend()

# Show the plot
plt.show() 
````

````python
#line plot for selling_price_window from above cell and correlation_list and other 
plt.plot(Selling_price_window, correlation_list, marker='s', linestyle='--', color='green', label='Line 2')

# Set labels and title
plt.xlabel('Selling price window')
plt.ylabel('Correlation')
plt.title('Line Graph for Two Lists')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)  # Add grid lines
plt.show()
````

## Question 5. How does the distribution of `salesprice` vary by make for cars made between 2010-2015?
- Use the 10 most common makes (your cleaned version of this column).  
- Create a graph that answers this question.

````python
#filtering the rows where column is in range between these years
filtered_df = df2[(df2['year'] >= 2010) & (df2['year'] <= 2015)]
filtered_df
````

````python
#value_counts() --> it helps in counting the individual value and giving top 10 highest frequency
most_common_names = filtered_df['make'].value_counts().nlargest(10)
#converting into dictionary
most_common_names.to_dict()
````

Sample narrative description.

````python
#filtering the makes where make is in above dictionary keys
filtered_makes = filtered_df[filtered_df['make'].isin(most_common_names.keys())]
filtered_makes
````

````python
# here we group the year and make columns and we are summing the sellingprice
result = filtered_makes.groupby(['year', 'make'])['sellingprice'].sum().reset_index()
print(result)
````

````python
#converting dict keys into list
makes_list = list(most_common_names.keys())
#sorting the list
makes_list = sorted(makes_list)
#storing selling price values in its respective years
_2010 = result.loc[0:9, 'sellingprice'].to_list()
_2011 = result.loc[10:19, 'sellingprice'].to_list()
_2012 = result.loc[20:29, 'sellingprice'].to_list()
_2013 = result.loc[30:39, 'sellingprice'].to_list()
_2014 = result.loc[40:49, 'sellingprice'].to_list()
_2015 = result.loc[50:59, 'sellingprice'].to_list()

plt.figure(figsize=(20,16))

bar_width = 0.1

# Create an array of x positions for the bars
x = np.arange(len(makes_list))

# Create the figure and axis objects
fig, ax = plt.subplots()


bar1 = ax.bar(x - bar_width, _2010, width=bar_width, label='2010')
bar2 = ax.bar(x, _2011, width=bar_width, label='2011')
bar3 = ax.bar(x + bar_width, _2012, width=bar_width, label='2012')
bar4 = ax.bar(x +(2*bar_width), _2013, width=bar_width, label='2013')
bar5 = ax.bar(x +(3*bar_width), _2014, width=bar_width, label='2014')
bar6 = ax.bar(x +(4*bar_width), _2015, width=bar_width, label='2015')
# Set the x-axis labels
ax.set_xticks(x)
ax.set_xticklabels(makes_list,rotation=45)

# Set labels and title
ax.set_xlabel('Car names')
ax.set_ylabel('Sellingprice')
ax.set_title('Multiple Bars for Each Car')

# Add a legend
ax.legend()

# Show the plot
plt.show()

````
